````python
You can create two `int` type variables, perform addition, subtraction, multiplication, and division operations on them, and then print the results in the specified format. Here's the code to do that:

```python
# Create two int variables
first_variable = 10
second_variable = 5

# Perform arithmetic operations and store the results
addition_result = first_variable + second_variable
subtraction_result = first_variable - second_variable
multiplication_result = first_variable * second_variable
division_result = first_variable / second_variable

# Print the results in the specified format
print(f"First variable is {first_variable} & second variable is {second_variable}.")
print(f"Addition: {first_variable} + {second_variable} = {addition_result}")
print(f"Subtraction: {first_variable} - {second_variable} = {subtraction_result}")
print(f"Multiplication: {first_variable} * {second_variable} = {multiplication_result}")
print(f"Division: {first_variable} / {second_variable} = {division_result}")
```

In this code:

- We create two `int` variables, `first_variable` and `second_variable`.
- We perform the addition, subtraction, multiplication, and division operations on these variables and store the results in separate variables (`addition_result`, `subtraction_result`, `multiplication_result`, and `division_result`).
- We use formatted string literals (f-strings) to print the results in the specified format, including variable values and the results of each operation.

When you run this code, it will produce the following output:

```
First variable is 10 & second variable is 5.
Addition: 10 + 5 = 15
Subtraction: 10 - 5 = 5
Multiplication: 10 * 5 = 50
Division: 10 / 5 = 2.0
```
````

````python
Q.2. What is the difference between the following operators:
(i) ‘/’ & ‘//’
(ii) ‘**’ & ‘^’
````

````python
(i) Difference between `/` and `//` operators:

- `/` (Division Operator): The `/` operator is used for regular division in Python. It returns the result of the division as a floating-point number (even if the inputs are integers).

   Example:
   ```python
   result = 7 / 3
   print(result)  # Output: 2.3333333333333335
   ```

- `//` (Floor Division Operator): The `//` operator is used for floor division in Python. It returns the largest integer less than or equal to the result of the division. The result is always an integer, and any fractional part is truncated.

   Example:
   ```python
   result = 7 // 3
   print(result)  # Output: 2
   ```

(ii) Difference between `**` and `^` operators:

- `**` (Exponentiation Operator): The `**` operator is used for exponentiation in Python. It raises the left operand to the power of the right operand.

   Example:
   ```python
   result = 2 ** 3
   print(result)  # Output: 8
   ```

- `^` (Bitwise XOR Operator): The `^` operator is used for bitwise XOR (exclusive OR) in Python. It performs a bitwise XOR operation between corresponding bits of two integers.

   Example:
   ```python
   result = 5 ^ 3
   print(result)  # Output: 6 (binary: 101 ^ 011 = 110)
   ```

In summary:

- `/` and `//` operators are both used for division, but `/` returns a floating-point result, while `//` returns an integer result (floor division).
- `**` is used for exponentiation, whereas `^` is used for bitwise XOR. They serve entirely different purposes in Python.
````

````python
Q.3. List the logical operators.
````

````python
In Python, there are three logical operators used to perform logical operations on Boolean values (True and False). These logical operators are:

1. **`and`**: The `and` operator returns `True` if both operands are `True`, and `False` otherwise.

   Example:
   ```python
   x = True
   y = False
   result = x and y
   print(result)  # Output: False
   ```

2. **`or`**: The `or` operator returns `True` if at least one of the operands is `True`. It returns `False` only if both operands are `False`.

   Example:
   ```python
   x = True
   y = False
   result = x or y
   print(result)  # Output: True
   ```

3. **`not`**: The `not` operator is a unary operator that returns the opposite Boolean value of its operand. If the operand is `True`, `not` returns `False`, and if the operand is `False`, `not` returns `True`.

   Example:
   ```python
   x = True
   result = not x
   print(result)  # Output: False
   ```

These logical operators are often used to combine or manipulate Boolean values and are fundamental in controlling the flow of programs through conditional statements and loops.
````

````python
Q.4. Explain right shift operator and left shift operator with examples.
````

````python
In Python, the right shift (`>>`) and left shift (`<<`) operators are used for bitwise shift operations on integers. These operators shift the bits of an integer left or right by a specified number of positions. Bitwise shift operations are often used for low-level bit manipulation, such as encoding and decoding data.

Here's an explanation of both operators with examples:

1. **Right Shift Operator (`>>`)**:
   - The right shift operator shifts the bits of an integer to the right by a specified number of positions.
   - It discards the rightmost bits and fills in the vacated positions on the left with either zeros or the sign bit, depending on whether the integer is signed or unsigned.
   - The syntax is: `x >> y`, where `x` is the integer to be shifted, and `y` is the number of positions to shift.

   Example 1 (right shift by 1 position):
   ```python
   x = 12  # Binary: 1100
   result = x >> 1
   print(result)  # Output: 6 (Binary: 0110)
   ```

   Example 2 (right shift by 2 positions with a negative integer):
   ```python
   x = -12  # Binary: 11111111111111111111111111110100 (Two's complement)
   result = x >> 2
   print(result)  # Output: -3 (Binary: 11111111111111111111111111111101)
   ```

2. **Left Shift Operator (`<<`)**:
   - The left shift operator shifts the bits of an integer to the left by a specified number of positions.
   - It discards the leftmost bits that exceed the size of the integer's representation and fills in the vacated positions on the right with zeros.
   - The syntax is: `x << y`, where `x` is the integer to be shifted, and `y` is the number of positions to shift.

   Example 1 (left shift by 1 position):
   ```python
   x = 6  # Binary: 0110
   result = x << 1
   print(result)  # Output: 12 (Binary: 1100)
   ```

   Example 2 (left shift by 3 positions with overflow):
   ```python
   x = 5  # Binary: 0101
   result = x << 3
   print(result)  # Output: 40 (Binary: 101000, leftmost bits are discarded)
   ```

Bitwise shift operations can be useful for optimizing code or working with binary representations of data. However, it's important to be aware of the potential overflow or underflow issues when shifting bits, especially when working with signed integers.
````
